#### <SQL로 영화 이름가져오기>

#### KOBIS 박스오피스 순위(집계기간 : 1962.11.03 ~ 2021.7.15)의 총 23403편 중, 첫 1000만 관객이 나온 2003년 12월 부터 2021.7.15까지 관객수가 100만명 이상인 영화의 수는 777 이다

```python
sql = """
SELECT
    movie_name
FROM
  `multicampus-semi-project-rbk.semi_project_rbk.boxoffice_rank_KOBIS`
WHERE
    release_date >= "2003-12-01"
    AND release_date < "2021-07-15"
    AND num_movie_audience >= 1000000
ORDER BY
    order_id
"""
df = client.query(sql).to_dataframe()
df
```

```python
movie_name_list = df["movie_name"].tolist()
```

```python
len(movie_name_list)
```

```python
movie_name_list
```

```python
#지나치게 빠른 크롤링으로 인해 크롤링이 막힐 경우를 대비해 time모듈 import
import time

#URL 패턴의 분석
url_template = "https://search.naver.com/search.naver?query="

failed_names = []
info_url_list = []

for movie_name in movie_name_list:
    try:
        url = f"{url_template}영화 {movie_name}"
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'lxml')
        #영화 + 영화명으로 검색후, '기본정보'로 이동
        info_url = soup.find("ul", class_="tab_list").find_all("li")[1].find("a").get("href").split("&query=")[-1]
        info_url_list.append((movie_name, info_url))
        print(movie_name, info_url)
        time.sleep(1)
    except Exception as e:
        print(f"{url} 러닝타임 정보가져오기에 실패 하였습니다", e)
        failed_names.append(movie_name)
```

```python
print(failed_names)
```

```python
runtime_list = []
for movie_name, info_url in info_url_list:
    try:
        url = url_template + info_url
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'lxml')
        runtime = int(soup.find("dl",class_="info").find_all("div")[4].find("dd").text.sub(r'분',''))
        runtime_list.append((movie_name, runtime))
        print(movie_name, runtime)
        time.sleep(1)
    except:
        print("실패", movie_name, info_url)
```

### 흥행의 기준 - 가설(11): 주연배우

#### 어떤 주연배우가 출연한 작품이 흥행 하였나? 

```python
# 주연배우별 흥행
```

```python
import time

url_template = "https://search.naver.com/search.naver?query="

lead_role_list=[]

for movie_name in movie_name_list:
    try:
        url = f"{url_template}영화 {movie_name} 주연배우"
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'lxml')
        
        lead_role_tag_list = soup.find_all("div", class_="title_box")
        role_names = []
        for lead_role_tag in lead_role_tag_list:
            role_name = lead_role_tag.find("a", class_="_text").text.strip()
            role_names.append(role_name)
        lead_role_list.append((movie_name, role_names))
        print(movie_name, role_names)
        time.sleep(0.5)    
    except Exception as e:
        print("실패",movie_name, e)

print(lead_role_list)
```

```python
import csv

with open("./data/lead_role_names.csv", "w", newline='') as fw:
    writer =csv.writer(fw)
    for i,lead_role in enumerate(lead_role_list): 
        if i == 0:
            writer.writerow(["movie_name","lead_role_names"])
        writer.writerow(lead_role)
```

```python
sql11 = """
SELECT
    rank.movie_name,
    lead_role.lead_role_names
FROM
  `multicampus-semi-project-rbk.semi_project_rbk.boxoffice_rank_KOBIS` AS rank
LEFT JOIN
  `multicampus-semi-project-rbk.semi_project_rbk.lead_role_names` AS lead_role
ON
    rank.movie_name = lead_role.movie_name
WHERE
    rank.release_date >= "2003-12-01"
    AND rank.release_date < "2023-11-14"
    AND rank.num_movie_audience >= 1000000
ORDER BY
    rank.order_id
"""    

df11 = client.query(sql11).to_dataframe()
df11

```

```python
new_list=[]
for lead_role in lead_role_list:
    new_list.append(lead_role) 
    
new_list
```

```python
plt.rc('font', family='Malgun Gothic')

plt.xticks(rotation=80)

sns.barplot(   
    x='lead_role_names',
    y='',
    data=df11
).set(
    title="영화 주연배우 별 흥행 순위"
)
```

### 흥행의 기준 - 가설(12): 각본가

#### 어떤 각본가가 쓴 작품이 흥행 하였나? 

```python
import time

url_template = "https://search.naver.com/search.naver?query="

screen_writer_list=[]

for movie_name in movie_name_list:
    try:
        url = f"{url_template}영화 {movie_name} 각본"
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'lxml')
        
        screen_writer_tag_list = soup.find("ul",class_="_panel").find_all("li")
        writer_names = []
        for screen_writer_tag in screen_writer_tag_list:
            writer_name = screen_writer_tag.find("a",class_='name').find("span",class_="_text").text
            writer_names.append(writer_name)
        screen_writer_list.append((movie_name, writer_names))
        print(movie_name,writer_names)
        time.sleep(1)    
    except Exception as e:
        print("실패",movie_name, e)

print(screen_writer_list)
```

```python
import csv

with open("./data/screen_writer_names.csv", "w", newline='') as fw:
    writer =csv.writer(fw)
    for i,screen_writer in enumerate(screen_writer_list): 
        if i == 0:
            writer.writerow(["movie_name","screen_writer_names"])
        writer.writerow(screen_writer)
```

```python
#주연배우 - OK

url_template = "https://search.naver.com/search.naver?query="
movie_name = "서울의 봄"

try:
    url = f"{url_template}영화 {movie_name} 주연배우"
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'lxml')

    lead_role_tag_list = soup.find_all("div", class_="title_box")
    role_names = []
    for lead_role_tag in lead_role_tag_list:
        role_name = lead_role_tag.find("a", class_="_text").text.strip()
        role_names.append(role_name)

except Exception as e:
    print("실패",movie_name, e)

print(role_names)
```
